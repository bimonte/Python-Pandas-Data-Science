### 05. Quais produtos são geralmente vendidos em conjunto (bundle)?

```python
VENDAS.head()
```

```python
BUNDLE = VENDAS[VENDAS["Order ID"].duplicated(keep = False)] # keep = False marca todas as duplicidades como True
BUNDLE
```

```python
BUNDLE['Combinacao'] = BUNDLE.groupby('Order ID').Product.transform(lambda prod: ''.join('|' + prod + '|'))
BUNDLE
```

```python
BUNDLE = BUNDLE.sort_values(by = ['Order ID', 'Product'])
```

```python
BUNDLE.groupby('Combinacao').count()['Order ID']
```

```python
BUNDLE.groupby('Combinacao').count()['Order ID'].sort_values(ascending = False)
```

```python
DF = pd.DataFrame()

for i in BUNDLE.Product.unique():
    
    for j in BUNDLE.Product.unique():
        
        a, b = sorted([i, j])
        
        a = "|" + a + "|"
        b = "|" + b + "|"
        
        bundle = a + b
        
        qty = BUNDLE[BUNDLE.Combinacao.map(lambda x : a in x and b in x)]["Order ID"].nunique()
        mean = BUNDLE[BUNDLE.Combinacao.map(lambda x : a in x and b in x)]["Price Each"].astype(float).mean()
            
        if (a != b and qty > 0): DF = DF.append( {"Bundle" : bundle, "Qty" : qty, "Mean" : mean} , ignore_index=True)
        
DF = DF.sort_values(by="Qty", ascending=False).drop_duplicates()
```
